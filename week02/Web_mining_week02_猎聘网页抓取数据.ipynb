```python
import pandas as pd
from requests_html import HTMLSession
payload = {
    "keyword":"自媒体",
   
}

session = HTMLSession()
r = session.get("https://www.liepin.com/zhaopin/?key=自媒体", params=payload)
```

```python
with open ("20春_Web数据挖掘_week02_猎聘自媒体数据抓取.html", encoding = "utf8", mode = "w") as fp:
    fp.write(r.html.html)
```
